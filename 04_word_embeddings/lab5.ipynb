Mikolov et al. (2013) propose the use of an offset vector method in order to answer analogy questions.
For example, if you want to find the concept X which satisfies the analogy “X is to China as London is
to England”, you need to find the concept closest to the point X in the vector space:

$$X = \vec{V}_{\text{China}} + \vec{V}_{\text{London}} - \vec{V}_{\text{England}}$$

```python
def  capital_predictor(train_capital, train_country):
    pairs_count = testtuples['capital-common-countries']
    predictions = {}

    for true_capital, des_country in pairs_count:
        if des_country == train_country:
            continue

        pred = mymodel.most_similar(
            positive=[des_country, train_capital],
            negative=[train_country],
            topn=1
        )

        predictions[des_country] = pred[0]

    return predictions
```

```python
def evaluate_one_pair(train_capital, train_country):

    pairs_count = testtuples['capital-common-countries']
    predictions = capital_predictor(train_capital, train_country)

    correct = 0
    

    for true_capital, country in pairs_count:
        if country == train_country:
            continue

        pred_capital, score = predictions[country]

        if pred_capital == true_capital:
            correct += 1

    return correct / (len(pairs_count)-1)
```

```python
evaluate_one_pair("London", "England")
```

```python
def average_accuracy():
    pairs_count = testtuples['capital-common-countries']
    accuracies = {}

    for train_capital, train_country in pairs_count:
        acc = evaluate_one_pair(train_capital, train_country)
        accuracies[(train_capital, train_country)] = acc

    avg_acc = sum(accuracies.values()) / len(accuracies)

    return avg_acc, accuracies
```

```python
avg_acc, accuracies = average_accuracy()

print("Average accuracy:", avg_acc)

for pair, acc in accuracies.items():
    print(pair, "->", acc)
```

Word2Vec sometimes ranks:

"London" very high (because it’s strongly related)

By filtering them out, you force the model to:

actually solve the analogy

These capitals:

appear less frequently

##### Deeper insight (MSc level)

We are implicitly changing:

$\hat{w} = \arg\max_{w \in V} \cos\left(\mathbf{v}_w, \mathbf{v}_{des} + \mathbf{v}_{cap} - \mathbf{v}_{country}\right)$

into:

$\hat{w} = \arg\max_{w \notin S} \cos(\cdot)$

where 

$S =$ forbidden words

👉 This is constrained decoding

Same idea used in:

beam search constraints
LLM decoding filters
structured prediction

```python
def  country_predictor(train_capital, train_country):
    pairs_count = testtuples['capital-common-countries']
    predictions = {}

    for true_capital, des_country in pairs_count:
        if true_capital == train_capital:
            continue

        pred = mymodel.most_similar(
            positive=[true_capital, train_country],
            negative=[train_capital],
            topn=10
        )

        predictions[true_capital] = pred

        # Filtering out
        for word, score in pred:
            if word not in {train_capital, train_country, true_capital}:
                predictions[true_capital] = (word, score)
                break

    return predictions
```

Is performance same, higher, or lower?

In practice, it is often lower when predicting country from capital.

Why?

- capitals are often strongly associated with their country, but
- country vectors are usually broader and noisier semantic objects
- some capital vectors are cleaner landmarks than country vectors
- embeddings often encode “Athens → Greece” less symmetrically than “Greece → Athens”

Also country names have more varied contexts:

- politics
- people
- language
- culture
- demonyms

while capitals often behave more like named entities.

So usually:

- capital prediction from country performs better
- country prediction from capital performs a bit worse
